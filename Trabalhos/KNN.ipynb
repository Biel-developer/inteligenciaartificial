```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

LINK = "https://docs.google.com/spreadsheets/d/1g1aQ61vijh6uHJuc8sijeBEMsoIQ2a5yLwUK04Wptlg/export?format=csv"

base = pd.read_csv(LINK)

nomes = {
    "Carimbo de data/hora": "data",
    "Você ficou gripado no ano passado ?": "gripe",
    "Você tomou vacina da gripe no ano passado?": "vacina",
    "  Você frequentou no ano passado,  semanalmente ambientes com muitas pessoas? (salas cheias, ônibus, eventos, etc.)  ": "aglomeracao",
    "  Você viajou no ano passado mais de 100 km de distância?  ": "viagem",
    "  Você tem alergia nas vias aéreas (rinite, sinusite, etc.)?  ": "alergia",
    "Quantas horas você dormiu em média por noite no ano passado?": "sono",
    "Você praticou atividade física no ano passado?": "atividade",
    "Você se alimentou de forma balanceada no ano passado?": "dieta",
    "Em média, quantas vezes você lavou as mãos por dia no ano passado?": "higiene",
    "Na sua percepção, o seu nível de estresse no ano passado foi:": "estresse"
}

base = base.rename(columns=nomes)
base = base.drop(columns=["data"])

base["estresse"] = base["estresse"].fillna(base["estresse"].median())

entrada = base.drop(columns=["gripe"])
saida = base["gripe"]

entrada = pd.get_dummies(entrada, drop_first=True)

encoder = LabelEncoder()
saida = encoder.fit_transform(saida)

X_tr, X_ts, y_tr, y_ts = train_test_split(
    entrada, saida, test_size=0.2, random_state=10
)

modelo_knn = KNeighborsClassifier(n_neighbors=5)
modelo_knn.fit(X_tr, y_tr)

previsoes = modelo_knn.predict(X_ts)

print("\n" + "="*50)
print(" RESULTADO DO KNN ")
print("="*50)
print(f"Acurácia: {accuracy_score(y_ts, previsoes) * 100:.2f}%")
print("\nRelatório:")
print(classification_report(y_ts, previsoes, target_names=encoder.classes_))
```
