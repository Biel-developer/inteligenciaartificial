```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

ID_PLANILHA = "15J01zSzxRu5FpS9ID_nvzO1K0QcKnim5opMBConedJQ"
ABA = "0"

link = f"https://docs.google.com/spreadsheets/d/{ID_PLANILHA}/export?format=tsv&gid={ABA}"

dados = pd.read_csv(link, sep="\t", decimal=",")
print("Colunas:", dados.columns.tolist())

colunas_x = ["peso_kg", "velocidade_kmh"]
coluna_y = "veiculo"

dados[colunas_x] = dados[colunas_x].apply(pd.to_numeric, errors="coerce")
dados[coluna_y] = dados[coluna_y].str.strip().str.lower()
dados = dados.dropna()

encoder = LabelEncoder()
y = encoder.fit_transform(dados[coluna_y])
X = dados[colunas_x].values

print("Classes:", encoder.classes_)

X_treino, X_teste, y_treino, y_teste = train_test_split(
    X, y, test_size=0.2, random_state=7, shuffle=True
)

print(f"Treino: {len(X_treino)} | Teste: {len(X_teste)}")

scaler = StandardScaler()
X_treino = scaler.fit_transform(X_treino)
X_teste = scaler.transform(X_teste)

modelo_knn = KNeighborsClassifier(n_neighbors=3)
modelo_knn.fit(X_treino, y_treino)
pred_knn = modelo_knn.predict(X_teste)

print("\n--- KNN ---")
print("Acurácia:", accuracy_score(y_teste, pred_knn))
print(confusion_matrix(y_teste, pred_knn))
print(classification_report(y_teste, pred_knn, target_names=encoder.classes_))

modelo_arvore = DecisionTreeClassifier(random_state=7)
modelo_arvore.fit(X_treino, y_treino)
pred_tree = modelo_arvore.predict(X_teste)

print("\n--- Árvore de Decisão ---")
print("Acurácia:", accuracy_score(y_teste, pred_tree))
print(confusion_matrix(y_teste, pred_tree))
print(classification_report(y_teste, pred_tree, target_names=encoder.classes_))
```
