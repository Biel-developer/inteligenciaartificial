```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, export_text

LINK_PLANILHA = "https://docs.google.com/spreadsheets/d/1g1aQ61vijh6uHJuc8sijeBEMsoIQ2a5yLwUK04Wptlg/export?format=csv"

dados = pd.read_csv(LINK_PLANILHA)

colunas = {
    "Carimbo de data/hora": "data",
    "Você ficou gripado no ano passado ?": "gripe",
    "Você tomou vacina da gripe no ano passado?": "vacina",
    "  Você frequentou no ano passado,  semanalmente ambientes com muitas pessoas? (salas cheias, ônibus, eventos, etc.)  ": "aglomeracao",
    "  Você viajou no ano passado mais de 100 km de distância?  ": "viagem",
    "  Você tem alergia nas vias aéreas (rinite, sinusite, etc.)?  ": "alergia",
    "Quantas horas você dormiu em média por noite no ano passado?": "sono",
    "Você praticou atividade física no ano passado?": "atividade",
    "Você se alimentou de forma balanceada no ano passado?": "dieta",
    "Em média, quantas vezes você lavou as mãos por dia no ano passado?": "higiene",
    "Na sua percepção, o seu nível de estresse no ano passado foi:": "estresse"
}

dados = dados.rename(columns=colunas)
dados = dados.drop(columns=["data"])

dados["estresse"] = dados["estresse"].fillna(dados["estresse"].median())

variaveis = dados.drop(columns=["gripe"])
alvo = dados["gripe"]

variaveis = pd.get_dummies(variaveis, drop_first=True)

codificador = LabelEncoder()
alvo = codificador.fit_transform(alvo)

X_treino, X_teste, y_treino, y_teste = train_test_split(
    variaveis, alvo, test_size=0.2, random_state=7
)

modelo = DecisionTreeClassifier(max_depth=3, random_state=7)
modelo.fit(X_treino, y_treino)

regras = export_text(modelo, feature_names=list(variaveis.columns))

print("\n" + "="*50)
print(" REGRAS GERADAS PELO MODELO ")
print("="*50)
print("<= 0.5 → Não | > 0.5 → Sim")
print("class 1 → Gripado | class 0 → Não gripado")
print("-"*50)
print(regras)
```
